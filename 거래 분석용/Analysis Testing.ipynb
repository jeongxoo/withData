```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = "JUNGNA_AIRPOD_TRADE_0914_1.csv"
b = "JUNGNA_AIRPOD_TRADE_0914_2.csv"
c = "JUNGNA_AIRPOD_TRADE_0915_1.csv"
d = "JUNGNA_AIRPOD_TRADE_0916_1.csv"
e = "JUNGNA_AIRPOD_TRADE_0916_2.csv"
abcde = [a,b,c,d,e]
temp = [pd.read_csv(name) for name in abcde]
```

```python
alls = pd.DataFrame(columns=["category", "nick_name", "cafe_ranking", "date", "current_status", "product_name","contact" , "price", "payment", "trade", "can_trust", "text", "imgs", "imgs_length"])
for i in range(len(temp)):
    for j in range(len(temp[i])):
        alls = alls.append(temp[i].iloc[j])
```

```python
alls
```

```python
cnt = 0
airpod_temp = pd.DataFrame(columns=["category", "nick_name", "cafe_ranking", "date", "current_status", "product_name","contact" , "price", "payment", "trade", "can_trust", "text", "imgs", "imgs_length"])
for i in range(len(temp)):
    if "에어팟프로" in temp["product_name"][i] or "에어팟 프로" in temp["product_name"][i]:
        airpod_temp = airpod_temp.append(temp.iloc[i])
        cnt += 1
        
```

```python
airpod_temp
```

```python
for i in range(len(airpod_temp)):
    price = airpod_temp["price"].iloc[i]
    price = price.replace(",", "")
    price = price.replace("원", "")
    airpod_temp["price"].iloc[i] = int(price)
```

```python
airpod_price = airpod_temp["price"]
airpod_p = [a.replace(",", "") for a in airpod_price]
airpod_p = [int(a.replace("원", "")) for a in airpod_p]
```

```python
airpod_dict = {}
for a in airpod_p:
    try:
        airpod_dict[a] += 1
    except:
        airpod_dict[a] = 1
```

```python
airpod_dict = sorted(airpod_dict.items())
```

```python
airpod_dict
```

```python
x, y = [a[0] for a in airpod_dict], [a[1] for a in airpod_dict]
```

```python
plt.plot(x, y)
plt.xlim(0,350000)
plt.show()
```
