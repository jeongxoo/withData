```python
import pandas as pd
```

```python
temp = pd.read_csv("airpod_bank.csv")
```

```python
temp = temp["BANK"]
```

```python
bank_dic = {}
for i in temp:
    j = i.split(" ")
    try:
        bank_dic[j[0]] += 1
    except:
        bank_dic[j[0]] = 1
```

```python
bank_dic = sorted(bank_dic.items())
```

```python
sum(bank_dic.values())
```

```python
bank_dic
```

```python
bank_dic.sort(key=lambda x:x[1], reverse=True)
```

```python
bank_percent = {}
for k,v in bank_dic.items():
    print(k, " -> ", (v/690) * 100, "%")
    bank_percent[k] = (v/690) * 100
```

```python
sorted(bank_percent.items(), key= lambda x:x[1], reverse=True )
```
