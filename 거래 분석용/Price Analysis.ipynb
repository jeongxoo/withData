```python
tt = pd.DataFrame(columns=["id", "DATE", "CATEGORY", "GOODS", "PLATFORM", "PRICE", "OUTLINE"])
```

```python
date_price = {"4":[],"5":[],"6":[], "7":[], "8":[]}
for i in range(len(tt)):
    days = str(tt["DATE"].iloc[i]).split(".")
    if len(days) == 1:
        continue
    if 4<=int(days[1]) <=8:
        if 150000 <= tt["PRICE"].iloc[i] <= 300000:
            date_price[days[1]].append(tt["PRICE"].iloc[i]) 
```

```python
check
```

```python
jungna_price = {"7":[], "8":[], "9":[]}
for i in range(len(check)):
    days = str(check["date"].iloc[i]).split(".")
    if len(days) == 1:
        continue
    if 6<=int(days[1]) <=9:
        price = check["price"].iloc[i]
        price = price.replace(",", "")
        price = price.replace("원", "")
        if 150000 <= int(price) <= 300000:
            jungna_price[str(int(days[1]))].append(int(price)) 
```
